```python
def hex_decimal_converter(value):
    if isinstance(value, str):
        if value.startswith("0x"):
            # Convert hexadecimal to decimal
            decimal_value = int(value, 16)
            return decimal_value
        else:
            # Convert decimal to hexadecimal
            hex_value = hex(int(value))
            return hex_value
    elif isinstance(value, int):
        # Convert decimal to hexadecimal
        hex_value = hex(value)
        return hex_value
    else:
        raise ValueError("Invalid input value. Must be a string or an integer.")
```

```python
import subprocess

def remove_password_from_pdf(input_file, output_file, password):
    command = ['qpdf', '--password={}'.format(password), '--decrypt', input_file, output_file]
    try:
        subprocess.run(command, check=True)
        print("Password removed successfully.")
    except subprocess.CalledProcessError as e:
        print("Failed to remove password:", e)

# Usage
input_file = '/Users/dev/Downloads/license_terms_xdCcI4ibbG.pdf'
output_file = '/Users/dev/Downloads/unlocked.pdf'
password = 'VnA7tbPx'
remove_password_from_pdf(input_file, output_file, password)
```

```python
import PyPDF2

def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)

        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            page_content = page.extract_text()
            print("Page", page_num + 1, "Content:")
            print(page_content)

# Usage
file_path = output_file
read_pdf(file_path)
```
